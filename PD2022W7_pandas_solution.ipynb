### Import pandas and files

```python
import pandas as pd
```

```python
file1 = 'PeopleData.xlsx'

people = pd.read_excel(file1, sheet_name='People')
leaders = pd.read_excel(file1, sheet_name='Leaders')
location = pd.read_excel(file1, sheet_name='Location')
goals = pd.read_excel(file1, sheet_name='Goals')
dates = pd.read_excel(file1, sheet_name='Date Dim')
```

### Create last_name, first_name fields for the agent and the leader

1. First create in people table
2. Second create in leaders table
3. Remove the extra first_name and last_name columns

```python
# Create field in people table
people['Agent Name'] = people['last_name'] + ', ' + people['first_name']
people
```

```python
#Create field in leaders table
leaders['Leader Name'] = leaders['last_name'] + ', ' + leaders['first_name']
leaders
```

```python
#Drop the first_name and last_name columns from people
people = people.drop('first_name', axis=1)
people = people.drop('last_name', axis=1)
people
```

```python
#Drop the first_name and last_name columns from leaders
leaders = leaders.drop('first_name', axis=1)
leaders = leaders.drop('last_name', axis=1)
leaders
```

### Join the people, leader, and location tables together

1. Rename selected columns to avoid confusion
2. Join people and leaders on agent_id
3. Join the result with location on location_id
4. Remove the unnecessary columns

```python
#### Rename columns to avoid confusion
people = people.rename(columns={'id' : 'agent_id', 'Leader 1' : 'leader_id', 'Location ID' :'location_id'})
people
```

```python
#Rename columns
leaders = leaders.rename(columns={'id' : 'leader_id'})
leaders
```

```python
#Rename columns
location = location.rename(columns={'Location ID' : 'location_id'})
location
```

```python
#Join people and leaders on leader_id
join = pd.merge(people, leaders, on='leader_id', how='left')
join
```

```python
#Join the result on location id
join = pd.merge(join, location, how='left', on='location_id')
join
```

```python
#Drop location ID
join = join.drop('location_id', axis=1)
join
```

### Input the monthly metrics data, clean it, and join it to the agent/location detail

1. Import the metric data. 
    - Passing sheet_name=None will import all the sheets. They will automatically be imported as a multi-index
2. Use the multi-index to assign a month column to the dataframe
3. Reset the multi-index to flatten
4. Use fillna to combine columns with similar information that are differently named
5. Drop extra columns
6. Join the agent/location data in joined with the monthly metrics in stats on agent_id

```python
#Import the monthly metrics data

file2 = 'MetricData2021.xlsx'
stats = pd.concat(pd.read_excel(file2, sheet_name=None))
stats
```

```python
# Use .get_level_values to set a month column
stats['Month'] = stats.index.get_level_values(0)
```

```python
#Reset the index to flatten
stats = stats.reset_index()
stats
```

```python
#Use fillna to combine columns

stats['Calls Offered'] = stats['Calls Offered'].fillna(stats['Offered'])
stats['Calls Not Answered'] = stats['Calls Not Answered'].fillna(stats['Not Answered'])
stats['Calls Answered'] = stats['Calls Answered'].fillna(stats['Answered'])
```

```python
#Drop extra columns

stats = stats[['AgentID', 'Calls Offered', 'Calls Not Answered', 'Calls Answered', 'Total Duration', 'Sentiment', 'Month']]
stats
```

```python
#Join the agent/location data with the stats on agent_id

joined = pd.merge(join, stats, left_on='agent_id', right_on='AgentID')
joined
```

### Calculate agent metrics

1. Calculate percent of offered calls that weren't answered (for each agent, each month)
2. Calculate the average call duration for each agent, each month
2. Calculate whether the sentiment score met the goal
3. Calculate whether the not answered rate met the goal

```python
joined.columns
```

```python
#Calculate the percent of offered calls that weren't answered

joined['Not Answered Rate'] = joined['Calls Not Answered']/joined['Calls Offered']
```

```python
#Calculate the average duration 

joined['Agent Avg Duration'] = joined['Total Duration']/joined['Calls Answered']
```

```python
#Calculate whether sentiment score met the goal

joined['Sentiment Goal Met'] = joined['Sentiment'].apply(lambda x: True if x >= 0 else False)
```

```python
#Calculate whether not answered rate met the goal

joined['Not Answered Goal Met'] = joined['Not Answered Rate'].apply(lambda x: True if x < .05 else False)
```
