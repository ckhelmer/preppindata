# Preppin Data Week 2: The Prep School - Birthday Cakes

### Import file
1. Import pandas
2. Import csv

```python
import pandas as pd
df = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
df
```

3. Remove unnecessary columns (optional step)

```python
df = df[['id', 'pupil first name', 'pupil last name', 'Date of Birth',]]
df.head()
```

### Format pupil's name in First Name Last Name format

```python
df['Pupil Name'] = df['pupil first name'] + ' ' + df['pupil last name']
```

### Create the date for the pupil's birthday in calendar year 2022
1. Import datetime for datetime handling
2. Format 'Date of Birth' as datetime

```python
import datetime as dt
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
```

3. Create separate columns to identify month and day of birthday

```python
df['month'] = df['Date of Birth'].dt.month
df['day'] = df['Date of Birth'].dt.day
```

4. Add a column for year (2022)

```python
df['year'] = 2022
```

5. Create a single column using the three columns created in steps 3 & 4

```python
df["This Year's Birthday"] = pd.to_datetime(dict(year=df.year, month=df.month, day=df.day))
df["This Year's Birthday"]
```

### Find the day of week the birthday falls on & change Saturdays and Sundays to Friday

1. Find weekday (unaltered)

```python
df['Cake Needed On'] = df["This Year's Birthday"].dt.day_name()
```

```python
df['Cake Needed On'].value_counts()
```

2. Replace Saturday and Sunday with Friday

```python
df['Cake Needed On'] = df['Cake Needed On'].replace(['Saturday', 'Sunday'], 'Friday')
```

```python
df['Cake Needed On'].value_counts()
```

### Find what month the pupil's birthday falls within

```python
df['Month Name'] = df["This Year's Birthday"].dt.month_name()
```

### Count how many birthdays there are on each weekday in each month
1. Create a groupby on Month Name and Cake Needed on, aggregated by count
    - Reset the index to produce a flat dataframe rather than a multi-indexed dataframe

```python
gb = df.groupby(['Month Name', 'Cake Needed On']).count().reset_index()
gb
```

2. Remove duplicate columns to return a single count
3. Rename field to reflect the fact that it is a count

```python
gb = gb[['Month Name', 'Cake Needed On', 'id']]
gb = gb.rename(columns={'id' : 'BDs per Weekday and Month'})
gb
```

4. Merge the original dataframe with the groupby on Month Name and Cake Needed On

```python
m = pd.merge(df, gb, on=['Month Name', 'Cake Needed On'], how='left')
m
```

### Output the data
1. Remove extra columns & Reorder

```python
m = m[['Pupil Name', 'Date of Birth', "This Year's Birthday", 'Month Name', 'Cake Needed On', 'BDs per Weekday and Month']]
m.head()
```

2. Output to csv

```python
m.to_csv('pandas_solution.csv', index=False)
```
