### Import libraries and files

```python
import pandas as pd
```

```python
#Declare the 
file = 'TC22 Input.xlsx'
```

```python
#Import each sheet separately using the sheet_name argument for pd.read_excel

registrations = pd.read_excel(file, sheet_name='Registrations')
registrations.head()
```

```python
sessions = pd.read_excel(file, sheet_name='Sessions')
sessions
```

```python
online = pd.read_excel(file, sheet_name='Online Attendees')
online.head()
```

```python
inperson = pd.read_excel(file, sheet_name='In Person Attendees')
inperson.head()
```

### Join the Session Names to the Registration information

```python
reg = pd.merge(registrations, sessions, on='Session ID', how='left')
reg
```

### Remove Session ID Column

```python
reg = reg.drop('Session ID', axis=1)
reg.head()
```

### Clean Online Column
1. View all entries using value counts
2. Clean using .replace()

```python
reg['Online/In Person'].value_counts()
```

```python
reg['Online/In Person'] = reg['Online/In Person'].replace(['Onlyne', 'ONLINE', 'Onlin'], 'Online')
reg['Online/In Person'] = reg['Online/In Person'].replace(['IN PERSON', 'In Persn', 'In Persoon', 'Im Person'], 'In Person')

reg['Online/In Person'].value_counts()
```

### Split registration data into in person and online information

```python
ip = reg[reg['Online/In Person'] == 'In Person'].copy()
ol = reg[reg['Online/In Person'] == 'Online'].copy()
```

### Join the In Person registration with the In Person Attendees to figure out who did NOT attend the workshop they registered for

1. Use the indicator argument to add a column detailing which dataframe contained the original information
2. Filter on 'left_only' to keep only items found in ip (registration data) but not in inperson (attendance data)

```python
person = pd.merge(ip, inperson, on=['Session', 'First Name', 'Last Name'], how='left', indicator=True)
person
```

```python
person['_merge'].value_counts()
```

```python
inperson_missed = person[person['_merge'] == 'left_only'].copy()
inperson_missed
```

### Join the Online registration with the Online Attendees to figure out who did NOT attend the workshop they registered for

1. Use the indicator argument to add a column detailing which dataframe contained the original information
2. Filter on 'left_only' to keep only items found in ol (registration data) but not in online (attendance data)

```python
ol.columns
```

```python
online.columns
```

```python
line = pd.merge(ol, online, on=['Session', 'Email'], how='left', indicator=True)
line
```

```python
line['_merge'].value_counts()
```

```python
online_missed = line[line['_merge'] == 'left_only'].copy()
online_missed
```

### Union the online_missed and inperson_missed dataframes

```python
missed = online_missed.append(inperson_missed)
missed
```

### Drop the merge column

```python
missed = missed.drop('_merge', axis=1)
missed
```

### Calculate the company name based on the email provided

1. Use str.split with '@' as the positional argument; take the 1st indexed result
    - this splits the email into two substrings one before the @ and the rest after; we're interested in what comes after the @, the domain name
2. Use str.split with '.' as the positional argument; take the 0th indexed result
    - this splits the remaining domain name into the company, and then the .com/.org/.gov, etc; we're interested in the company
3. You can chain all of these functions to perform all calculations in a single line

```python
missed['Company'] = missed['Email'].str.split('@').str[1].str.split('.').str[0]
missed['Company']
```

### Calculate how many sessions each enrollee didn't attend

1. Create a groupby on 'Email' which functions as a unique key.
2. Pass .Session (the column name) as an argument to return only those values
3. use .nunique() as the aggregating function for the groupby
4. Sort values (not necessary, but useful for checks)
5. Reset index to create a new dataframe

```python
missed_gb = missed.groupby(by=['Email']).Session.nunique().sort_values(ascending=False).reset_index()
missed_gb
```

### Calculate how many sessions each attendee signed up for

1. Use the modified registrations table (reg) to count all sessions, not just the missed ones
2. Create a groupby on 'Email' which functions as a unique key.
3. Pass .Session (the column name) as an argument to return only those values
4. use .nunique() as the aggregating function for the groupby
5. Sort values (not necessary, but useful for checks)
6. Reset index to create a new dataframe

```python
reg_gb = reg.groupby(by=['Email']).Session.nunique().sort_values(ascending=False).reset_index()
reg_gb
```

### Join the two groupby's on email

Use a left join to return only information from people who missed sessions

```python
gb = pd.merge(missed_gb, reg_gb, on='Email', how='left', suffixes=['_missed', '_registered'])
gb
```

### Calculate the percentage of sessions missed for the remaining people

Round to two decimal places

```python
gb['Not Attended %'] = round((gb.Session_missed/gb.Session_registered)*100, 2)
gb['Not Attended %']
```

### Join the gb dataframe with the missed dataframe to get the company, name, and other information

1. Use a left join to retain all information from the missed table

```python
full = pd.merge(missed, gb, on='Email', how='left')
full
```

### Remove extra columns

```python
full.columns
```

```python
full = full.drop(['Session_missed', 'Session_registered'], axis=1)
full
```
