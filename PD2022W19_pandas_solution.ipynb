### Week 19: C&BSCo Good Sales but Wrong Sizes

### Import libraries and file

```python
import pandas as pd
```

```python
file = 'PD 2022 Wk 19 Input.xlsx'
```

```python
product = pd.read_excel(file, sheet_name='Product Set')
product
```

```python
sales = pd.read_excel(file, sheet_name='Sales')
sales
```

```python
size = pd.read_excel(file, sheet_name='Size Table')
size
```

### Change the Size ID to an actual size value in the Sales Table

Join the Sales Table to the Size Table on Size = Size ID

```python
join = pd.merge(sales, size, left_on='Size', right_on='Size ID')
join
```

```python
#Remove extra columns and rename for clarity

join = join.drop(['Size_x', 'Size ID'], axis=1)
join = join.rename(columns={'Size_y' : 'Sales Size'})

join
```

### Link the Product code to the sales table to provide the scent information

1. The Product Code from the Product Table and the Product from the Sales Table don't match; all 'S' characters can be removed with str.replace()
2. Rename the Size column from product code to prevent confusion 
3. Join the 'join' table (sales/sizes) to the product table on Product Code

```python
#Replace 'S' in Product Code
product['Product Code'] = product['Product Code'].str.replace('S', '')
product
```

```python
#Rename the Size column in the product table
product = product.rename(columns={'Size' : 'True Size'})
product
```

```python
#Join the join table with the product table
join2 = pd.merge(join, product, left_on='Product', right_on='Product Code')
join2
```

### Create an output that contains the products sold with correct sizes

1. Create an indicator column to indicate whether the Sales Sizes and True Sizes match
    - You have to convert both columns to strings rather than objects to make this work
2. Keep only results where indicator column = True
3. Output to csv

```python
#Convert string columns
join2['True Size'] = join2['True Size'].astype('string')
join2['Sales Size'] = join2['Sales Size'].astype('string')
join2
```

```python
#Create indicator column
join2['Indicator'] = join2['Sales Size'] == join2['True Size']
join2
```

```python
#Remove extra columns
join2 = join2.drop(['Product Code', 'Sales Size', 'Indicator'], axis=1)
join2
```

```python
#Keep only correct sizes (Indicator = True)
correct = join2[join2['Indicator'] == True]
correct
```

```python
#Export to csv
correct.to_csv('correct_orders.csv', index=False)
```

### Create another dataset to show all products sold with incorrect sizes
1. Aggregate the data using groupby to show each product sold, scent, and size with a count of sales incorrectly recorded

```python
gb = join2.groupby(by=['Product', 'True Size', 'Scent']).count().reset_index()
gb
```

```python
#Rename columns for clarity
gb = gb.rename(columns={'Store' : 'Sales with the wrong size', 'True Size' : 'Product Size'})
gb
```

```python
#Export to csv
gb.to_csv('wrong_orders.csv', index=False)
```
