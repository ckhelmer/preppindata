### 1. Import pandas and the file

```python
import pandas as pd
```

```python
df = pd.read_csv('2022W18 Input.csv')
df
```

### 2. Pivot the data (wide-to-long) so that each value can be viewed on its own line

Use df.melt to reshape the data using the id_var of 'Region'

```python
df = df.melt(id_vars=['Region'])
df
```

### 3. Split the resulting 'variable' column into fields describing bike type, date, and measure type

Use .str.split() with the expand=True argument to produce 3 separate columns; assign those column names via double-brackets

```python
df[['Bike Type', 'Date', 'Measure Type']] = df['variable'].str.split('___', expand=True)

df
```

Remove the original column as it is no longer necessary

```python
df = df.drop('variable', axis=1)
df
```

### 4. Create a field for Sales and Profit

1. Reshape the dataframe using df.pivot to put the profit and sales information in separate columns
2. Assign the profit and sales information to the original dataframe. (You can do this because the indexes still match.)
3. To produce the final dataframe with no null values in the Profit and Sales Columns
    1. Create two separate dataframes via dropna; one for sales, one for profit
    2. Remove unnecessary columns (sales, measure type, and value from profit, and profit, measure type, and value from sales) from each dataframe
    3. Merge the dataframes together on the remaining columns (bike type, date, region)

```python
reshaped = df.pivot(columns='Measure Type', values=['value']).reset_index()
reshaped
```

```python
#Assign profit and sales information from pivot to original dataframe
df['Profit'] = reshaped.value.Profit
df['Sales'] = reshaped.value.Sales
df
```

```python
#Split resulting dataframe into two to remove the nulls
p = df.dropna(subset=['Profit'])
s = df.dropna(subset=['Sales'])
p
```

```python
#Remove extra columns to prevent duplication in result
p = p.drop(['Sales', 'Measure Type', 'value'], axis=1)
s = s.drop(['Profit', 'Measure Type', 'value'], axis=1)
```

```python
#Join dataframes on Region, Bike Type, and Date to produce the un-pivoted dataframe
m = pd.merge(p, s, on=['Region', 'Bike Type', 'Date'])
m
```

### 5. Convert Date column to a pandas datetime using the format argument to specify

```python
m['Date'] = pd.to_datetime(m['Date'], format='%b_%y')
m
```

### 6. Export to csv

```python
m.to_csv('pandas_solution.csv', index='False')
```
