### 1. Import libraries and files

```python
import pandas as pd
```

```python
#Import the data using read_excel
file = 'Menu Input.xlsx'

orders = pd.read_excel(file, sheet_name='Orders')
lookup = pd.read_excel(file, sheet_name='Lookup Table')
```

```python
orders.head()
```

```python
lookup.head()
```

### 2. Perform basic data cleaning of orders table

1. Rename the columns for ease of processing
2. Drop unnecessary columns

```python
#Rename columns
orders = orders.rename(columns={'Unnamed: 1' : 'Carls Selections', 'Unnamed: 4' : 'Toms Selections', 
                                'Unnamed: 7' : 'Jennys Selections', 'Unnamed: 10' : 'Jons Selections'})
orders
                       
```

```python
#Drop extra columns
orders = orders.drop(['Unnamed: 2', 'Unnamed: 5', 'Unnamed: 8'], axis=1)
orders
```

### 3. Split the orders table into different courses, and add the courses as a column

1. Split each course into different columns
2. Assign each course's value in its own dataframe
3. Union the dataframes

```python
#Split into different courses
starters = orders.iloc[1:4].copy()
mains = orders.iloc[5:9].copy()
dessert = orders.iloc[10:].copy()
```

```python
starters
```

```python
mains
```

```python
dessert
```

```python
#Assign course values
starters['Course'] = 'Starter'
mains['Course'] = 'Main'
dessert['Course'] = 'Dessert'
```

```python
#Union course table

orders = starters.append(mains)
orders = orders.append(dessert)
orders
```

### 4. Reshape data to obtain a column for guest, course, and dish

For each guest:
1. Reshape the orders table using pd.melt() to have the course as the value, and the selection as the variable
2. Drop values which have not been selected (nans)
3. Remove selection column

Then, union the guest tables. Rename columns as necessary.

```python
#Melt carl's data, using Carl's Selections and Course as the ID, and 'Carl' (the actual dish) as the value
carl = pd.melt(orders, id_vars=['Carls Selections', 'Course'], value_vars=['Carl'])
carl
```

```python
#Drop nans to leave only selections
carl = carl.dropna(subset=['Carls Selections'])
carl
```

```python
#Drop selections column because it is unnecessary
carl = carl.drop('Carls Selections', axis=1)
carl
```

```python
#Repeat with the other three guests
tom = pd.melt(orders, id_vars=['Toms Selections', 'Course'], value_vars=['Tom'])
tom = tom.dropna(subset=['Toms Selections'])
tom = tom.drop('Toms Selections', axis=1)
tom
```

```python
jenny = pd.melt(orders, id_vars=['Jennys Selections', 'Course'], value_vars=['Jenny'])
jenny = jenny.dropna(subset=['Jennys Selections'])
jenny = jenny.drop('Jennys Selections', axis=1)
jenny
```

```python
jon = pd.melt(orders, id_vars=['Jons Selections', 'Course'], value_vars=['Jonathan'])
jon = jon.dropna(subset=['Jons Selections'])
jon = jon.drop('Jons Selections', axis=1)
jon
```

```python
#Append all 4 dataframes

orders = carl.append(tom)
orders = orders.append(jenny)
orders = orders.append(jon)
orders
```

```python
#Rename the columns
orders = orders.rename(columns={'variable' : 'Guest', 'value' : 'Dish'})
orders
```

###  5. Find each dish's ID

Join the lookup table to the orders tables to find each dish's ID

```python
#Join lookup table to orders table

full = pd.merge(orders, lookup, on='Dish')
full
```

### 6. Export the data to csv

```python
full.to_csv('pandas_solution.csv', index=False)
```
