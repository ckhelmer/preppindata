### House of Games Winners

### Import pandas and file

```python
import pandas as pd
import numpy as np
```

```python
file = "Richard Osman's House of Games - Episode Guide - Players.csv"
df = pd.read_csv(file)
df
```

### Cleanup

1. Drop unnecessary rows by index
2. Change column names for clarity
3. Drop unnecessary columns

```python
#Drop key rows from bottom

df = df.iloc[:300, :]
df
```

```python
#Change column names
df = df.rename(columns={'Ser.' : 'Series', 'Wk.': 'Week', 'T' : 'Tu', 'T.1' : 'Th', 
                        'Total' : 'Score', 'Week' : 'Points', 'Week.1' : 'Rank', 
                        'M.1' : 'M_R' , 'T.2' : 'Tu_R', 'W.1' : 'W_R', 'T.3' : 'Th_R', 'F.1' : 'F_R'})
```

```python
df = df[['Player', 'Series', 'Week', 'M', 'Tu', 'W', 'Th', 'F','M_R', 'Tu_R', 'W_R', 'Th_R', 'F_R','Score', 'Points', 'Rank']]
df
```

### Remove Series with Null Values or Preceded by N (N1, N2, etc.)

```python
df = df[~df['Series'].str.contains('N')].copy()
```

```python
#Convert Rank to Numeric dtype

df['Rank'] = df['Rank'].str.slice(stop=1).astype('int64')
df['M_R'] = df['M_R'].str.slice(stop=1).astype('int64')
df['Tu_R'] = df['Tu_R'].str.slice(stop=1).astype('int64')
df['W_R'] = df['W_R'].str.slice(stop=1).astype('int64')
df['Th_R'] = df['Th_R'].str.slice(stop=1).astype('int64')
df['F_R'] = df['F_R'].str.slice(stop=1).astype('int64')
```

### Calculate points without double points Friday

1. Write function to use with .apply() function
2. .apply() function

```python
def no_double_points_friday(x):
    
    score = 0
    
    for each in [x['M_R'], x['Tu_R'], x['W_R'], x['Th_R'], x['F_R']]:
            
        if each == 1:
            score += 4
        elif each == 2:
            score += 3
        elif each == 3:
            score += 2
        else:
            score += 1
        
    return score
```

```python
df['Points without Double Points Friday'] = df.apply(no_double_points_friday, axis=1)
df
```

### Rank the players based on hypothetical standings after no double points on Friday

1. Use a chained groupby and rank function to rank players by episode
    - 'dense' argument allows for ties to be recorded as 1, 1, 3 rather than 1.5, 1.5, 3

```python
df['Rank_Points_without_double_Friday'] = df.groupby(['Series', 'Week'])['Points without Double Points Friday'].rank('dense', ascending=False).astype('int64')
df
```

### Rank players to assign a winner if calculated using only total score

```python
#Rank players by total score
df['Rank by Total Score'] = df.groupby(['Series', 'Week'])['Score'].rank('dense', ascending=False).astype('int64')
```

### Calculate points if the Score on Friday was doubled

1. Calculate new Friday score and save as column
2. Calculate new rankings based on this field

```python
#Calculate score with double score fridays

df['Double Score Friday'] = (df['F']*2) + df['M'] + df['Tu'] + df['W'] + df['Th']
```

```python
#Rank based on double score Friday

df['DSF_Rank'] = df.groupby(['Series', 'Week'])['Double Score Friday'].rank('dense', ascending=False).astype('int64')
```

### Determine if winners have changed by any of the previous methods

1. Determine original winners (by series/week)
2. Figure out if winners would have changed given new scoring methodology
    - Use .apply() function to compare 'Rank' column (original) with new methodology columns (x3)
    - Drop episodes where original winner was a tie; this prevents the join from working as it should
3. Pare down information to contain only Series/Week/New Winner information
4. Join to original data

```python
# Find original winners by filtering on rank == 1
og = df[df['Rank'] == 1].copy()
og
```

```python
#Write functions to compare original rank to new rank
def change_dpf(x):
    
    if x['Rank'] != x['Rank_Points_without_double_Friday']:
        return 'Change'
    else:
        return 'No Change'
    
def change_dsf(x):
    
    if x['Rank'] != x['DSF_Rank']:
        return 'Change'
    else:
        return 'No Change'
    
def change_total_score(x):
    
    if x['Rank'] != x['Rank by Total Score']:
        return 'Change'
    else:
        return 'No Change'
```

```python
#Create change in winner fields using df.apply()

og['Change in winner with no double points Friday'] = og.apply(change_dpf, axis=1)
og['Change in winner based on Score'] = og.apply(change_total_score, axis=1)
og['Change in winner if Double Score Friday'] = og.apply(change_dsf, axis=1)
```

```python
# Save tie episodes to their own dataframe

ties = og[og.duplicated(subset=['Series', 'Week'], keep=False)].reset_index()
ties
```

```python
#None of these episodes would have resulted in a tie; manually set the three indicator fields to 'Changes'

ties.iloc[:, 22:] = 'Change'
ties
```

```python
#Remove episodes where there originally was a tie

og = og[~og.duplicated(subset=['Series', 'Week'], keep=False)]
```

```python
#Pare down og dataframes to reunite changes with original data
og = og[['Series', 'Week','Change in winner with no double points Friday',
       'Change in winner based on Score',
       'Change in winner if Double Score Friday']]
og
```

```python
#Do the same thing with ties
ties = ties[['Series', 'Week','Change in winner with no double points Friday',
       'Change in winner based on Score',
       'Change in winner if Double Score Friday']]

#Drop the duplicates to prevent a double key
ties = ties.drop_duplicates()
ties
```

```python
#Append Ties and nontie winners

og = og.append(ties)
og
```

```python
#Merge the dataframes
df2 = pd.merge(df, og, on=['Series', 'Week'])
df2
```

### Remove Unnecessary Fields
